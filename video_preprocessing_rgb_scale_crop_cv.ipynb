# Image Crop

### 테스트

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = './man2.jpg'

img_bgr = cv2.imread(file_path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap='gray')
```

```python
body_cascade = cv2.CascadeClassifier('./opencv/haarcascade_fullbody.xml')

body = body_cascade.detectMultiScale(img_gray, scaleFactor=1.001, minNeighbors=10)
```

```python
body
```

```python
for (x,y,w,h) in body :         
    cv2.rectangle(img_rgb, (x,y), (x+w,y+h), (0,0,255), 3)
    
plt.imshow(img_rgb)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = './NP/Assault/Assault002_x264.npy'

arr_img = np.load(file_path)
```

```python
cv2.imwrite('./test.jpg', arr_img[0])
img_rgb = cv2.imread('test.jpg')

plt.figure(figsize=(12,12))
plt.imshow(img_rgb)
plt.show()
```

```python
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(12,12))
plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
arr = np.load(file_path)

arr = arr[0] # 첫 번째 프레임

plt.imshow(arr)
plt.show()
```

```python
body_cascade = cv2.CascadeClassifier('./opencv/haarcascade_fullbody.xml')

body = body_cascade.detectMultiScale(img_gray, scaleFactor=1.001, minNeighbors=10)

body
```

detectMultiScale(image, objects, scaleFactor, minNeighbors, flags, minSize, maxSize)
- image : 입력 영상. CV_8U 깊이의 행렬
- objects : (출력) 검출된 객체의 사각형 좌표 정보
- scaleFactor : 검색 윈도우 확대 비율. 1보다 커야 합니다.
- minNeighbors : 검출 영역으로 선택하기 위한 최소 검출 횟수
- flags : 현재 사용되지 않습니다.
- minSize : 검출할 객체의 최소 크기
- maxSize : 검출할 객체의 최대 크기

```python
for (x,y,w,h) in body:
    cv2.rectangle(img_rgb, (x,y), (x+w,y+h), (0,0,255), 2)    
    
plt.figure(figsize=(12,12))
plt.imshow(img_rgb)
plt.show()
```
